```python
# --- 1. CONFIGURACIÓN DEL ENTORNO ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Configuración de visualización
pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 6)
warnings.filterwarnings('ignore') # Omitir advertencias de versiones futuras
```

```python
# --- 2. PROCESAMIENTO DE ACTIVOS FINANCIEROS E INFLACIÓN ---

# A) RENTA VARIABLE (Euro Stoxx 50 Total Return)
# Fuente: Bloomberg/Investing. Se toma el cierre trimestral.
df_stoxx = pd.read_excel("eurostoxx_gr.xlsx", header=2) # header=2 según tu archivo
df_stoxx = df_stoxx[["Date", "SX5T Index"]].dropna()
df_stoxx["Date"] = pd.to_datetime(df_stoxx["Date"])
df_stoxx.set_index("Date", inplace=True)
# Transformación a cierre trimestral y cálculo de rentabilidad nominal
r_stoxx_q = df_stoxx["SX5T Index"].resample("QE").last().pct_change()

# B) RENTA FIJA SOBERANA (REX Performance Index)
# Fuente: Deutsche Börse. Índice de retorno total.
df_rex = pd.read_excel("rex_performance.xlsx", header=1) # header=1 según tu archivo
df_rex = df_rex[["Date", "REXP Index"]].dropna()
df_rex["Date"] = pd.to_datetime(df_rex["Date"])
df_rex.set_index("Date", inplace=True)
r_rex_q = df_rex["REXP Index"].resample("QE").last().pct_change()

# C) ACTIVO LIBRE DE RIESGO (Bund 1Y)
# Fuente: Investing. Yield anualizado.
df_bund = pd.read_excel("germany_1y.xlsx", sheet_name="alemán a 1 año Datos Históricos")
df_bund = df_bund[["Fecha", "Último"]].rename(columns={"Fecha": "Date", "Último": "Yield_pct"})
df_bund["Date"] = pd.to_datetime(df_bund["Date"])
df_bund.set_index("Date", inplace=True)
df_bund.sort_index(inplace=True)
# Cierre trimestral del yield y conversión a tanto por uno
bund_yield_q = df_bund["Yield_pct"].resample("QE").last() / 100
# Desanualización geométrica (Ref: Ecuación [119])
r_f_q = (1 + bund_yield_q)**(1/4) - 1

# D) INFLACIÓN (HICP España)
# Carga específica para tu formato de Excel (filas 8 y 10)
df_hicp_raw = pd.read_excel("hipc.xlsx", sheet_name="Sheet 1", header=None)
dates = df_hicp_raw.iloc[8, 1:].values
values = df_hicp_raw.iloc[10, 1:].values
df_hipc = pd.DataFrame({"Date": dates, "HICP": values}).dropna()
df_hipc["Date"] = pd.to_datetime(df_hipc["Date"]) + pd.offsets.MonthEnd(0) # Fin de mes
df_hipc["HICP"] = pd.to_numeric(df_hipc["HICP"])
df_hipc.set_index("Date", inplace=True)
# Cálculo de inflación trimestral (Ref: Ecuación [123])
df_infl_q = df_hipc["HICP"].resample("QE").last().pct_change()
df_infl_q.name = "Inflacion_trimestral"
```

```python
# --- 3. CONSTRUCCIÓN DE LA RENTABILIDAD DE LA VIVIENDA ---

# A) Carga de Datos
# Precio (MITMA)
df_precio = pd.read_excel("vivienda_total_nacional_limpio.xlsx")
df_precio["Fecha"] = pd.to_datetime(df_precio["Fecha"])
df_precio = df_precio.set_index("Fecha").sort_index()
serie_precio = df_precio["Precio"]

# Alquiler (Banco de España)
df_alq = pd.read_excel("alquiler_bruto.xlsx")
df_alq["Fecha"] = pd.to_datetime(df_alq["Fecha"])
df_alq = df_alq.dropna().set_index("Fecha").sort_index()
serie_yield_anual = df_alq["Alquiler_bruto"] / 100

# B) Cálculo de Componentes
# 1. Plusvalía (Revalorización de precio) (Ref: Ecuación [79])
r_precio_q = serie_precio.pct_change()

# 2. Rentabilidad por Alquiler (Ref: Ecuación [84] y [98])
# Desanualización geométrica
r_alq_bruta_q = (1 + serie_yield_anual)**(1/4) - 1
# Ajuste por costes y vacancia (25%)
COSTES_VACANCIA = 0.25
r_alq_neta_q = (1 - COSTES_VACANCIA) * r_alq_bruta_q

# C) Rentabilidad Total (Total Return) (Ref: Ecuación [104])
# Unión temporal de ambas series
df_vivienda = pd.concat([r_precio_q, r_alq_neta_q], axis=1)
df_vivienda.columns = ["r_precio", "r_alquiler_neto"]

# Fórmula multiplicativa compuesta
df_vivienda["Rentabilidad_Total_Vivienda"] = (
    (1 + df_vivienda["r_precio"]) * (1 + df_vivienda["r_alquiler_neto"])
) - 1

# Eliminar el primer trimestre (donde precio es NaN)
df_vivienda_final = df_vivienda.dropna()
```

```python
# --- 4. CONSOLIDACIÓN DE DATOS (MERGE) ---

# Unimos todas las series alineándolas por su índice trimestral (Fecha)
df_consolidado = pd.concat([
    df_vivienda_final["Rentabilidad_Total_Vivienda"], # Vivienda
    r_stoxx_q,                                        # Bolsa
    r_rex_q,                                          # Bonos
    r_f_q,                                            # Risk Free
    df_infl_q                                         # Inflación
], axis=1)

# Renombrar columnas para claridad académica
df_consolidado.columns = [
    "Nominal_Vivienda",
    "Nominal_RV_Europa",
    "Nominal_RF_Soberana",
    "Nominal_RiskFree",
    "Inflacion"
]

# Eliminar filas con datos faltantes (Intersection de fechas)
df_consolidado = df_consolidado.dropna()

print(f"Base de datos consolidada creada. Total trimestres: {len(df_consolidado)}")
df_consolidado.head()
```

```python
# --- 5. CÁLCULO DE RENTABILIDADES REALES ---
# Metodología: Ecuación de Fisher exacta (1+r_real) = (1+r_nom)/(1+pi) - 1
# Ref: Ecuación [125] y [145] del Marco Metodológico.

df_real = pd.DataFrame(index=df_consolidado.index)
activos = ["Vivienda", "RV_Europa", "RF_Soberana", "RiskFree"]

for activo in activos:
    col_nom = f"Nominal_{activo}"
    col_real = f"Real_{activo}"
    # Aplicación de la fórmula
    df_real[col_real] = (1 + df_consolidado[col_nom]) / (1 + df_consolidado["Inflacion"]) - 1

print("Transformación a términos reales completada.")
df_real.tail()
```

```python
# --- 6. ANÁLISIS DE RENDIMIENTO Y RIESGO (RESULTADOS) ---

# A) Cálculo de Primas de Riesgo (Excess Returns)
# Ref: Ecuación [153]
df_excess = pd.DataFrame(index=df_real.index)
df_excess["Exceso_Vivienda"] = df_real["Real_Vivienda"] - df_real["Real_RiskFree"]
df_excess["Exceso_RV_Europa"] = df_real["Real_RV_Europa"] - df_real["Real_RiskFree"]
df_excess["Exceso_RF_Soberana"] = df_real["Real_RF_Soberana"] - df_real["Real_RiskFree"]

# B) Tabla de Estadísticos (Media, Volatilidad, Sharpe)
resultados = pd.DataFrame(index=["Vivienda", "RV_Europa", "RF_Soberana"])

# 1. Rentabilidad Real Media (Trimestral)
resultados["Retorno_Real_Medio_Q"] = df_real[["Real_Vivienda", "Real_RV_Europa", "Real_RF_Soberana"]].mean().values

# 2. Volatilidad (Desviación Estándar) (Ref: Ecuación [161])
resultados["Volatilidad_Q"] = df_real[["Real_Vivienda", "Real_RV_Europa", "Real_RF_Soberana"]].std().values

# 3. Prima de Riesgo Media (Ref: Ecuación [158])
resultados["Exceso_Retorno_Medio_Q"] = df_excess.mean().values

# 4. Ratio de Sharpe (Ref: Ecuación [165])
resultados["Sharpe_Ratio_Q"] = resultados["Exceso_Retorno_Medio_Q"] / resultados["Volatilidad_Q"]

# C) Exportación y Visualización
print("\n=== RESULTADOS EMPÍRICOS FINALES (1996-2025) ===")
display(resultados.round(4))

archivo_salida = "TFG_Resultados_Definitivos.xlsx"
with pd.ExcelWriter(archivo_salida) as writer:
    df_consolidado.to_excel(writer, sheet_name="1_Nominales")
    df_real.to_excel(writer, sheet_name="2_Reales")
    df_excess.to_excel(writer, sheet_name="3_Excesos")
    resultados.to_excel(writer, sheet_name="4_Metricas_Finales")
```

```python
# --- GRÁFICO 1: RIQUEZA ACUMULADA (BASE 100) ---
import matplotlib.pyplot as plt

# Estilo
plt.style.use('seaborn-v0_8-whitegrid')
colors = {'Vivienda': '#2E86C1', 'RV_Europa': '#E74C3C', 'RF_Soberana': '#27AE60', 'RiskFree': '#95A5A6'}

# Cálculo (Si ya lo hiciste antes no pasa nada por repetirlo aquí)
wealth_index = (1 + df_real).cumprod() * 100
wealth_index.iloc[0] = 100

plt.figure(figsize=(10, 6)) 
for col in ['Real_Vivienda', 'Real_RV_Europa', 'Real_RF_Soberana', 'Real_RiskFree']:
    label = col.replace('Real_', '')
    plt.plot(wealth_index.index, wealth_index[col], label=label, color=colors.get(label, 'black'), linewidth=2)

plt.title('Evolución de la Riqueza Acumulada (1996-2025)', fontsize=14, fontweight='bold')
plt.ylabel('Valor de la Inversión (Base 100)', fontsize=12)
plt.legend(frameon=True, fontsize=11)
plt.grid(True, linestyle='--', alpha=0.7)

# Guardar y Mostrar
plt.tight_layout()
plt.savefig('Grafico_1_Riqueza.png', dpi=300)
plt.show()
```

```python
# --- GRÁFICO 2: MAPA RIESGO - RENTABILIDAD ---

plt.figure(figsize=(10, 6))

# Datos
x = resultados['Volatilidad_Q']
y = resultados['Retorno_Real_Medio_Q']
labels = resultados.index

# Dibujar puntos
plt.scatter(x, y, s=200, c=[colors.get(l, 'gray') for l in labels], edgecolors='black', alpha=0.9)

# Etiquetas de texto
for i, txt in enumerate(labels):
    plt.annotate(txt, (x[i], y[i]), xytext=(10, 5), textcoords='offset points', fontsize=12, fontweight='bold')

plt.title('Eficiencia: Riesgo vs Rentabilidad (Datos Trimestrales)', fontsize=14, fontweight='bold')
plt.xlabel('Riesgo (Volatilidad Trimestral)', fontsize=12)
plt.ylabel('Retorno (Rentabilidad Media Trimestral)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Guardar y Mostrar
plt.tight_layout()
plt.savefig('Grafico_2_Riesgo_Retorno.png', dpi=300)
plt.show()
```

```python
# --- GRÁFICO 3: DRAWDOWN (CAÍDAS DESDE MÁXIMOS) ---
# CORRECCIÓN DE ERROR: Forzamos que los datos sean 'float' (números) explícitamente

# Cálculo
rolling_max = wealth_index.cummax()
drawdown = (wealth_index - rolling_max) / rolling_max

# --- LÍNEA CLAVE PARA SOLUCIONAR TU ERROR ---
drawdown = drawdown.astype(float)
# --------------------------------------------

plt.figure(figsize=(10, 6))

# Vivienda
plt.plot(drawdown.index, drawdown['Real_Vivienda'], label='Vivienda', color=colors['Vivienda'], linewidth=2)
# El error salía aquí abajo. Ahora con .astype(float) funcionará:
plt.fill_between(drawdown.index, drawdown['Real_Vivienda'], 0, color=colors['Vivienda'], alpha=0.1)

# Bolsa (para comparar)
plt.plot(drawdown.index, drawdown['Real_RV_Europa'], label='RV Europa', color=colors['RV_Europa'], linewidth=1, alpha=0.6)

plt.title('Perfil de Riesgo: Caídas desde Máximos (Drawdown)', fontsize=14, fontweight='bold')
plt.ylabel('% Caída', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)

# Guardar y Mostrar
plt.tight_layout()
plt.savefig('Grafico_3_Drawdown.png', dpi=300)
plt.show()
```
